```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None, index_col = None)
train_set.head()
```

```python
train_set[1].unique()
```

```python
# what is the count of all these above labels i need to know 
train_set[1].value_counts()
```

```python
# only selecting categorical features 
columns = [1, 3, 5, 6, 7, 8, 9, 13]
```

```python
train_set=train_set[columns]
```

```python
train_set.head()
```

```python
train_set.columns=['Employment', 'Degree', 'Status', 'Designation','Family_job', 'Race', 'Sex', 'Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature, ":", len(train_set[feature].unique()), 'labels')
```

```python
country_map = train_set['Country'].value_counts().to_dict()
```

```python
country_map
```

```python
train_set['Country'] = train_set['Country'].map(country_map)
```

```python
train_set.head(20)
```

### Advantages
1. Easy to use implement.
2. Not increasing any feature space. 


### Disadvantage
1. It will provide same weight if the frequencies are same, so model will not differentiate. 

### Target Guided Ordinal Encoding 

1. Ordering the labels according to the target.
2. Replace the labels by the joint probability of being 1 or 0 

```python
import pandas as pd
df= pd.read_csv(r"C:\Users\User\Desktop\Titanic.csv", usecols=['Cabin', 'Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace=True)
```

```python
df.head(10)
```

```python
df['Cabin']= df['Cabin'].astype(str).str[0] # Converted first into string then grab out first letter which will give cabin row.
```

```python
df.head()
```

```python
# in above M letter shows 'Missing' Value not a Row. 
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean() # Survived percentage cabin wise except "M" which is a missing percentage.
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels
```

```python
ordinal_labels2= {k:i for i,k in enumerate(ordinal_labels, 0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels'] = df['Cabin'].map(ordinal_labels2)
```

### Probability Ratio Encoding 

```python
1. Probability of survived based on cabin -- categorical feature 
2. Probability of not survived --- 1 - pr(survived)
3. pr(survived)/pr(not survived)
4. Dictionary to map cabin with probability 
5. replace with the categorical feature 
```

```python
import pandas as pd 
```

```python
df= pd.read_csv(r"C:\Users\User\Desktop\Titanic.csv", usecols=['Cabin', 'Survived'])
df.head()
```

```python
# Replacing Nan Value with some other keywords 
df['Cabin'].fillna('Missing', inplace=True)
```

```python
df.head(10)
```

```python
df['Cabin'].unique()
```

```python
# Just take first letter of unique values 
df['Cabin']= df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean() # Survival cabin wise 
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
# what is the probability that person has died 
prob_df['Died'] = 1- prob_df['Survived']
```

```python
prob_df.head()
```

```python
# probability ratio encoding i.e survived / death 
prob_df['Probability_ratio'] = prob_df['Survived']/prob_df['Died']
```

```python
prob_df.head()
```

```python
probability_encoded=prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded']= df['Cabin'].map(probability_encoded)
```
